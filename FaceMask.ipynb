```python
# Single-cell: Build DB from known folder, recognize faces in a group image (works well with partial masks)
# Run this in Google Colab.
!pip install -q insightface==0.7.3 opencv-python-headless==4.7.0.68 matplotlib

import os, cv2, numpy as np
from pathlib import Path
from google.colab import drive
from insightface.app import FaceAnalysis
from base64 import b64encode
import matplotlib.pyplot as plt

# ----- USER-CHANGEABLE PATHS -----
KNOWN_DIR = '/content/drive/MyDrive/known'   # folder with known person images (label from filename)
INPUT_IMAGE_PATH = '/content/drive/MyDrive/known'  # the 4-person image you uploaded (or '/content/photo.jpg')
OUTPUT_PATH = '/content/drive/MyDrive/known/result_identified.jpg'
# ---------------------------------

# mount drive
drive.mount('/content/drive', force_remount=True)

# initialize insightface (detection + recognition). ctx_id=0 uses GPU if available, -1 = CPU
app = FaceAnalysis(allowed_modules=['detection','recognition'])
try:
    app.prepare(ctx_id=0, det_size=(640,640))
except Exception as e:
    # fallback to CPU if GPU provider not available
    print("GPU init failed, falling back to CPU:", e)
    app.prepare(ctx_id=-1, det_size=(640,640))

def l2_normalize(x, eps=1e-10):
    x = np.array(x, dtype=np.float32)
    norm = np.linalg.norm(x)
    if norm < eps:
        return x
    return x / norm

def cosine_sim(a, b):
    a = l2_normalize(a)
    b = l2_normalize(b)
    return np.dot(a, b)

# 1) Build database of embeddings from KNOWN_DIR
db_embeddings = []
db_labels = []

p = Path(KNOWN_DIR)
if not p.exists():
    raise FileNotFoundError(f"Known directory not found: {KNOWN_DIR}")

for img_file in sorted(p.iterdir()):
    if img_file.suffix.lower() not in ('.jpg', '.jpeg', '.png'):
        continue
    # use filename (without ext) as label
    label = img_file.stem
    img = cv2.imread(str(img_file))
    if img is None:
        print("Could not read", img_file); continue
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    faces = app.get(img_rgb)
    if len(faces) == 0:
        print("No face found in known image:", img_file)
        continue
    # pick the largest face found (in case of group pictures accidentally in known folder)
    faces = sorted(faces, key=lambda f: (f.bbox[2]-f.bbox[0])*(f.bbox[3]-f.bbox[1]), reverse=True)
    emb = faces[0].embedding
    db_embeddings.append(l2_normalize(emb))
    db_labels.append(label)
print(f"Loaded {len(db_labels)} known faces.")

# 2) Load input image (group or captured photo)
img = cv2.imread(INPUT_IMAGE_PATH)
if img is None:
    raise FileNotFoundError(f"Input image not found: {INPUT_IMAGE_PATH}")
orig = img.copy()
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# 3) Detect faces and compute embeddings
faces = app.get(img_rgb)
print("Detected faces:", len(faces))

# 4) Match each detected face to DB using cosine similarity
THRESHOLD = 0.48   # tune from 0.4..0.55. Lower = stricter. For masked faces, slightly higher may be needed.
results = []
for face in faces:
    x1, y1, x2, y2 = face.bbox.astype(int)
    emb = l2_normalize(face.embedding)
    best_sim = -1.0
    best_label = "Unknown"
    for db_emb, db_label in zip(db_embeddings, db_labels):
        sim = cosine_sim(emb, db_emb)
        if sim > best_sim:
            best_sim = sim
            best_label = db_label
    # decide match or unknown
    if best_sim >= THRESHOLD:
        name = f"{best_label} ({best_sim:.2f})"
    else:
        name = f"Unknown ({best_sim:.2f})"
    results.append((x1,y1,x2,y2,name,best_sim))

# 5) Draw results
for (x1,y1,x2,y2,name,sim) in results:
    cv2.rectangle(orig, (x1,y1), (x2,y2), (0,200,0), 2)
    # put filled rect for text background
    ((w,_), _) = cv2.getTextSize(name, cv2.FONT_HERSHEY_SIMPLEX, 0.6, 1)
    y_text = max(y1-10, 10)
    cv2.rectangle(orig, (x1, y_text-20), (x1 + w + 6, y_text+4), (0,200,0), -1)
    cv2.putText(orig, name, (x1+3, y_text), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0,0,0), 1, cv2.LINE_AA)

# 6) Save + show
cv2.imwrite(OUTPUT_PATH, orig)
print("Saved result to:", OUTPUT_PATH)

# display inline
plt.figure(figsize=(10,10))
plt.axis('off')
plt.imshow(cv2.cvtColor(orig, cv2.COLOR_BGR2RGB))
plt.show()
```

#How It Works
The system extracts a face from the video frame.
It converts the face into a numeric representation called an embedding.
It compares this embedding with stored embeddings of known persons.
The closest match becomes the predicted identity.
A confidence score is also shown based on similarity.
Even with small data, this method can work very well in controlled CCTV setups.

1. Input Frame (CCTV/Webcam)
2. MTCNN detects faces
3. Each face cropped and aligned
4. Convert to tensor
5. Pass through InceptionResNetV1
6. Generate embedding
7. Compare with stored embeddings
8. Compute minimum Euclidean distance
9. Convert to confidence score
10. Display Name + Confidence on screen

```python
#Building Embedding Database
import os
import torch
import pickle
from PIL import Image
from torchvision import transforms
from facenet_pytorch import InceptionResnetV1

# ---------------------------
# 1. SET DATASET PATH
# ---------------------------
dataset_path = "/content/drive/My Drive/face_dataset"
output_pickle = "/content/drive/My Drive/face_embeddings.pkl"

# ---------------------------
# 2. LOAD FACENET MODEL
# ---------------------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = InceptionResnetV1(pretrained='vggface2').eval().to(device)    #.eval() = inference mode

# ---------------------------
# 3. IMAGE PREPROCESSOR
# ---------------------------
transform = transforms.Compose([
    transforms.Resize((160, 160)),
    transforms.ToTensor(),
])    #   # FaceNet requires 160x160 input

# ---------------------------
# 4. START BUILDING Embedding DATABASE
# ---------------------------
embedding_database = {}

for person in os.listdir(dataset_path):
    person_path = os.path.join(dataset_path, person)

    # If not a folder (like hidden files), skip

    if not os.path.isdir(person_path):
        continue

    print(f"\nProcessing person:", person)

    embeddings = []

    # look inside "mask" and "unmask"
    for subfolder in ["mask", "unmask"]:
        sub_path = os.path.join(person_path, subfolder)

        if not os.path.isdir(sub_path):
            continue

        for img_name in os.listdir(sub_path):
            img_path = os.path.join(sub_path, img_name)

            try:
                img = Image.open(img_path).convert("RGB")
                img_tensor = transform(img).unsqueeze(0).to(device)

                with torch.no_grad():
                    embedding = model(img_tensor).cpu().numpy()[0]

                embeddings.append(embedding)

                print("✔ Embedded:", img_path)

            except Exception as e:
                print(f"⚠ Error processing {img_path}:", e)

# Save Embeddings

    if len(embeddings) > 0:
        embedding_database[person] = embeddings
        print(f"🎯 Stored {len(embeddings)} embeddings for:", person)
    else:
        print(f"❌ No embeddings stored for:", person)

# ---------------------------
# 5. SAVE TO DISK(Load Stored Embeddings)
# ---------------------------
with open(output_pickle, "wb") as f:
    pickle.dump(embedding_database, f)

print("\n🎉 All embeddings saved at:", output_pickle)
```

```python


import pickle

with open("/content/drive/My Drive/face_embeddings.pkl", "rb") as f:
    database = pickle.load(f)

print("Loaded names:", list(database.keys()))

# ===============================
#  Initialize Detection + Embedding Models
# ===============================
device = 'cuda' if torch.cuda.is_available() else 'cpu'

# MTCNN → Detect face and get bounding boxes
detector = MTCNN(keep_all=True, device=device)

# FaceNet → Generate embedding for face crop
embedder = InceptionResnetV1(pretrained='vggface2').eval().to(device)



# ===============================
# FUNCTION TO GET EMBEDDING OF CROPPED FACE
# ===============================
# def get_embedding(pil_img):
#     img_tensor = embedder(
#         torch.unsqueeze(
#             torch.tensor(np.array(pil_img)).permute(2,0,1) / 255.0,
#             0
#         ).float().to(device)
#     )
#     return img_tensor.detach().cpu().numpy().reshape(-1)
def get_embedding(pil_img):
    img = pil_img.resize((160, 160))     # Resize to FaceNet standard
    img = np.array(img).astype(np.float32) / 255.0   # Normalize
    img = torch.tensor(img).permute(2,0,1).unsqueeze(0).to(device)
    with torch.no_grad():
        emb = embedder(img)
    return emb.cpu().numpy().reshape(-1)


# ===============================
# IDENTIFICATION FUNCTION
# ===============================
def identify_face(embedding, threshold=1.1):
    best_match = None
    best_score = 100    # Lowest distance = best match


    for name, emb_list in database.items():
        for ref_emb in emb_list:
            dist = np.linalg.norm(embedding - ref_emb)
            if dist < best_score:
                best_score = dist
                best_match = name

     # If within threshold → recognized, else unknown
    if best_score < threshold:
        return best_match, best_score
    return "Unknown", best_score



# ===============================
# RUN ON TEST IMAGE
# ===============================
img_path = "photo.jpg"
img = Image.open(img_path)

boxes, _ = detector.detect(img)

# Convert to CV2 for drawing
img_draw = cv2.cvtColor(np.array(img), cv2.COLOR_RGB2BGR)

results = []

if boxes is None:
    print("❌ No face detected.")
else:
    for i, box in enumerate(boxes):
        x1, y1, x2, y2 = map(int, box)

        face_crop = img.crop((x1, y1, x2, y2))
        embedding = get_embedding(face_crop)
        name, score = identify_face(embedding)


        results.append((name, score))

        # Draw box
        cv2.rectangle(img_draw, (x1, y1), (x2, y2), (0,255,0), 2)

        # Write name
        cv2.putText(
            img_draw,
            name,
            (x1, y1 - 10),
            cv2.FONT_HERSHEY_SIMPLEX,
            0.8,
            (0,255,0),
            2
        )


    # Save output image
    cv2.imwrite("output_identified.jpg", img_draw)

    print("✔ Output image saved: output_identified.jpg")
    print("\n===== DETECTION SUMMARY =====")
    for i, (name, dist) in enumerate(results, start=1):
        print(f"Face {i} -> {name}   (distance={dist:.4f})")


```
