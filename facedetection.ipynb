```python
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
eye_cascade  = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')
mouth_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_smile.xml')
vedio_c=cv2.VideoCapture(0)
while(True):
    _,frame=vedio_c.read()
    grey =cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    face=face_cascade.detectMultiScale(grey,1.1,5)
    for (x,y,w,h) in face :
        roi_grey=grey[y:y+h,x:x+w]
        roi_color=frame[y:y+h,x:x+w]
        eyes=eye_cascade.detectMultiScale(roi_grey,1.1,5)
        for (ex,ey,ew,eh) in eyes:
            cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)
            cv2.putText(roi_color, "Eye", (ex, ey-5),cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 1)
        mouth_roi_gray = roi_grey[int(h/2):, :]        #Focus only on the lower half of the face where the mouth is likely
        mouth = mouth_cascade.detectMultiScale(mouth_roi_gray, 1.3, 15)

        for (mx, my, mw, mh) in mouth:
            my = my + int(h/2)  # Adjust because ROI starts mid-face
            cv2.rectangle(roi_color, (mx, my), (mx+mw, my+mh), (0,0,255), 2)
            cv2.putText(roi_color, "Mouth", (mx, my-5),cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 1)

    cv2.imshow("face dtection",frame)
    if cv2.waitKey(1) & 0xFF == 27:
        break
vedio_c.release()
cv2.destroyAllWindows()
```

```python
def parse_wider_annotations(txt_path, img_root):
    annotations = []
    with open(txt_path, 'r') as f:
        lines = f.readlines()

    i = 0
    while i < len(lines):
        img_path = lines[i].strip()
        i += 1

        try:
            num_faces = int(lines[i].strip())
        except ValueError:
            print(f"Skipping malformed entry at line {i}: {lines[i]}")
            i += 1
            continue

        i += 1
        faces = []
        for _ in range(num_faces):
            face_data = list(map(float, lines[i].strip().split()))
            faces.append(face_data)
            i += 1

        annotations.append({
            'img_path': os.path.join(img_root, img_path),
            'faces': faces
        })

    return annotations
train_annotations = parse_wider_annotations(r"C:\Users\dixit\Downloads\wider_face_split\wider_face_split\wider_face_train_bbx_gt.txt", r"C:\Users\dixit\Downloads\WIDER_train\WIDER_train\images")
val_annotations = parse_wider_annotations(r"C:\Users\dixit\Downloads\wider_face_split\wider_face_split\wider_face_val_bbx_gt.txt", r"C:\Users\dixit\Downloads\WIDER_val\WIDER_val\images")

print("Train samples:", len(train_annotations))
print("Val samples:", len(val_annotations))
print("First train sample:", train_annotations[0])




# wider_face_split = r"C:\Users\dixit\Downloads\wider_face_split\train.txt"
# val_ann_file = r"C:\Users\dixit\Downloads\wider_face_split\val.txt"

# train_img_dir = r"C:\Users\dixit\Downloads\WIDER_train\WIDER_train\images"
# val_img_dir = r"C:\Users\dixit\Downloads\WIDER_val\WIDER_val\images"


```

```python
import cv2

def convert_to_yolo(annotations, output_dir):
    os.makedirs(output_dir, exist_ok=True)
    for ann in annotations:
        img_path = ann['img_path']
        faces = ann['faces']

        # Load image to get dimensions
        img = cv2.imread(img_path)
        if img is None:
            print(f"Warning: could not read {img_path}")
            continue
        h, w = img.shape[:2]

        # Create YOLO annotation file
        txt_name = os.path.splitext(os.path.basename(img_path))[0] + ".txt"
        txt_path = os.path.join(output_dir, txt_name)

        with open(txt_path, "w") as f:
            for face in faces:
                x, y, bw, bh = face[:4]  # first 4 values are bbox
                # Convert to YOLO format (normalized)
                x_center = (x + bw / 2) / w
                y_center = (y + bh / 2) / h
                bw_norm = bw / w
                bh_norm = bh / h
                f.write(f"0 {x_center} {y_center} {bw_norm} {bh_norm}\n")

# ✅ Convert train and val sets
convert_to_yolo(train_annotations, "/content/wider_face/yolo_labels/train")
convert_to_yolo(val_annotations, "/content/wider_face/yolo_labels/val")

print("YOLO conversion done! Labels saved in /content/wider_face/yolo_labels/")
```
